```python
def get_daily_stats(data,lis,key):
    dic={}
    
    data = data[data["Package"]==lis[key]]
    
    result={}
    for i in range(data.shape[0]):
        if str(data["timestamp"].iloc[i].date()) not in result:
            result[str(data["timestamp"].iloc[i].date())]=data["UT"].iloc[i]
        else:
            result[str(data["timestamp"].iloc[i].date())] += data["UT"].iloc[i]
    del dic
    return result
```
